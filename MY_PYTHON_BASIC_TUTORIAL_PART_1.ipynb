### Working with more than one arguments at the same time:
#### Problem statement: Suppose, The DC charecter Flash one day discovered that he has the ability to pick up Thors hammer and can throw it in the speed of light to disintegrate it into energy, As, according to the famous equation:
> *E = mc2*
#### Let us assume that mass of thors hammer is 10000 kg and speed of light is 2.9 * 10^-8 m/s
#### Calculate the energy release???

```python
#lets define a function Release_Energy, containing the input values for mass(m)and speed of light(c)
#fomulla is defined bu E, so to get the result we need to return E:

def Released_Energy(m,c):
    E = m * c ** 2
    print("Mass =", m, "kg")
    print("speed of light =", c, "miles/sec")
    print("Released energy in Joules:")
    return E
    
```

### Simple linear regression analysis :
#### - It is also known as Ordinary Least Square (OLS) method. <br> - Least Square is also known as the minimum squared error or minSSE. <br> (lower the error meands better explanatory potwntial of the data in the regression model) <br> -  with the help of Calculus and linear algebra, <br> The ideal slope and intercepts are determine computationally closest to all the data points. 



#### Three majoy terms falls under the ANOVA framework are:
#### - sum of squares total (SST) <br>(squsred differences between observed and its mean) <br>

#### - sum of squares regression (SSR) <br> (sum of the differences between the predicted value and the mean of the dependent variables) <br> 

#### - sum of sruares error (SSE) <br> (sum of the squares of the error - lower the value better) <br>

### Relation between all three:
#### $SST = SSR + SSE$

#### - Its not about the best fitting line anymore, <br>
